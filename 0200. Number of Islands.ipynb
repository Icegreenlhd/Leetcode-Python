```python
dx, dy = [-1, 1, 0, 0], [0, 0, -1, 1]
class Solution(object):
    def numIslands(self, grid):
        """
        :type grid: List[List[str]]
        :rtype: int
        """
        if len(grid) <= 0 or len(grid[0]) <= 0: return 0
        self.grid, self.m, self.n = grid, len(grid), len(grid[0])
        count = 0
        for i in range(self.m):
            for j in range(self.n):
                if self.grid[i][j] == '1':
                    count += 1
                    self._floodfill(i, j)
        return count

    def _floodfill(self, i, j):
        self.grid[i][j] = 0
        for k in range(4):
            x, y = i+dx[k], j+dy[k]
            if self.m>x>=0 and self.n>y>=0:
                if self.grid[x][y] == '1':
                    self._floodfill(x, y)
```

```python
class DisjoinSet(object):
    def __init__(self, grid):
        m, n = len(grid), len(grid[0])
        self.roots = [-1] * (m*n)
        self.count = 0
        for i in range(m):
            for j in range(n):
                if grid[i][j] == '1':
                    self.roots[i*n+j] = i*n+j
                    self.count += 1
    
    def findroot(self, i):
        root = i
        while root != self.roots[root]:
            root = self.roots[root]
        while i != self.roots[i]:
            tmp = self.roots[i]
            self.roots[i] = root
            i = tmp
        return root

    def connected(self, p, q):
        return self.findroot(p) == self.findroot(q)
    
    def union(self, p, q):
        qroot = self.findroot(p)
        proot = self.findroot(q)
        if proot != qroot:
            self.roots[proot] = qroot
            self.count -= 1

class Solution(object):
    def numIslands(self, grid):
        """
        :type grid: List[List[str]]
        :rtype: int
        """
        if len(grid) <= 0 or len(grid[0]) <= 0: return 0 
        ds = DisjoinSet(grid)
        directions = [(1, 0), (0, 1)]
        m, n = len(grid), len(grid[0])

        for i in range(m):
            for j in range(n):
                if grid[i][j] == '1':
                    for d in directions:
                        x, y = i+d[0], j+d[1]
                        if 0<=x<m and 0<=y<n and grid[x][y] == '1':
                            ds.union(i*n+j, x*n+y)
        return ds.count
```

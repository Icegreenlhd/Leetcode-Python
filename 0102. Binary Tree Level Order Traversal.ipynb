```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def levelOrder(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        if not root: return []
        queue = []
        queue.append(root)
        result = []
        while len(queue) > 0:
            level_queue = queue[:]
            queue = []
            result.append([])
            while len(level_queue) > 0:
                q = level_queue.pop(0)
                result[-1].append(q.val)
                if q.left != None:
                    queue.append(q.left)
                if q.right != None:
                    queue.append(q.right)
        return result
        
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def levelOrder(self, root):
        """
        :type root: TreeNode
        :rtype: List[List[int]]
        """
        if not root: return []
        result = []
        self.dfs(root, 1, result)
        return result
    
    def dfs(self, root, level, result):
        if root != None:
            if len(result) < level:
                result.append([])
            result[level-1].append(root.val)
            self.dfs(root.left, level+1, result)
            self.dfs(root.right, level+1, result)
```

```python
class TreeNode(object):
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

def init_bst(tree):
    t = tree.copy()
    root = TreeNode(t.pop(0))
    tree_list = [root]
    while len(t) > 0:
        size = len(tree_list)
        for i in range(size):
            temp = tree_list.pop(0)
            if temp != None:
                left = t.pop(0) if len(t) > 0 else None
                temp.left  = TreeNode(left) if left else None
                right = t.pop(0) if len(t) > 0 else None
                temp.right  = TreeNode(right) if right else None
                tree_list.append(temp.left)
                tree_list.append(temp.right)
    return root

import unittest


class LeetcodeTest(unittest.TestCase):
    solution = getattr(Solution(), dir(Solution)[-1])

    def test_case1(self):
        test_input = [3,9,20,None,None,15,7]
        test_input = init_bst(test_input)
        test_output = [[3],[9,20],[15,7]]
        self.assertEqual(self.solution(test_input), test_output)

unittest.main(argv=['ignored', '-v'], exit=False)
```

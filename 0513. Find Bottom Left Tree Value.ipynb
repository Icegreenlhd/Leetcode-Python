```python
# BFS
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution(object):
    def findBottomLeftValue(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        if root is None: return -1
        stack = [root]
        lm = -1
        while len(stack) > 0:
            st = []
            lm = stack[0].val
            while len(stack) > 0:
                r = stack.pop(0)
                if r.left != None:
                    st.append(r.left)
                if r.right != None:
                    st.append(r.right)
            stack = st
        return lm
```

```python
# DFS 递归
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution(object):
    def findBottomLeftValue(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        self.n = 0
        self.lm = -1
        self._dfs(root, 1)
        return self.lm
    
    def _dfs(self, root, level):
        if root is None:
            return
        self._dfs(root.left, level+1)
        if level > self.n:
            self.n = level
            self.lm = root.val
        self._dfs(root.right, level+1)
        

```

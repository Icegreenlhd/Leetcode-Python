```python
# 核心覆盖点数
import random
class Solution(object):

    def __init__(self, rects):
        """
        :type rects: List[List[int]]
        """
        self.rects = rects
        self.sum = [0]
        for a, b, x, y in rects:
            self.sum.append(self.sum[-1]+(x-a+1)*(y-b+1))

    def pick(self):
        """
        :rtype: List[int]
        """
        w = random.randint(1, self.sum[-1])
        # 二分查找
        l, r = 0, len(self.sum)
        while l <= r:
            mid = (l+r) // 2
            if self.sum[mid] >= w:
                r = mid-1
            else:
                l = mid
            
        rect = self.rects[r]
        x = random.randint(rect[0], rect[2])
        y = random.randint(rect[1], rect[3])
        return [x, y]



# Your Solution object will be instantiated and called as such:
# obj = Solution(rects)
# param_1 = obj.pick()
```

```python
# 最基础的二叉树维护操作
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def deleteNode(self, root, key):
        """
        :type root: TreeNode
        :type key: int
        :rtype: TreeNode
        """
        if root is None:
            return None
        if root.val > key:
            root.left = self.deleteNode(root.left, key)
        elif root.val < key:
            root.right = self.deleteNode(root.right, key)
        elif root.left is None or root.right is None:
            root = root.left if root.left else root.right
        else:
            successor = root.right
            while successor.left:
                successor = successor.left
            successor.right = self.deleteNode(root.right, successor.val)
            successor.left = root.left
            return successor
        return root
            
            
```

```python
# 最基础的二叉树维护操作
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def deleteNode(self, root, key):
        """
        :type root: TreeNode
        :type key: int
        :rtype: TreeNode
        """
        cur, parent = root, None
        while cur and cur.val != key:
            parent = cur
            cur = cur.left if cur.val > key else cur.right
        if cur:
            cur = self.deletecur(cur)
        if parent is None:
            return cur
        if parent.left and parent.left.val == key:
            parent.left = cur
        else:
            parent.right = cur
        return root
            
    def deletecur(self, root):
        if root.right is None and root.left is None:
            root = None
            return root
        if root.left is None:
            return root.right
        if root.right is None:
            return root.left
        p, q = root.right, root
        while p.left:
            q = p
            p = p.left
        if q.val == root.val:
            q.right = p.right
        else:
            q.left = p.right
        p.right = root.right
        p.left = root.left
        root = p
        return root
```
